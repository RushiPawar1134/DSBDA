```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('iris.csv')
```

```python
df.hist()
```

```python
def graph(y):
     sns.boxplot(x='variety',y=y,data=df)

plt.subplot(221)
graph('sepal.width')

plt.subplot(222)
graph('sepal.length')

plt.subplot(223)
graph('petal.width')

plt.subplot(224)
graph('petal.length')

plt.show()
```

```python
Q1=np.percentile(df['sepal.width'],25,interpolation='midpoint')
```

```python
print(Q1)
```

```python
Q3=np.percentile(df['sepal.width'],75,interpolation='midpoint')
```

```python
IQR=Q3-Q1
```

```python
df.shape
```

```python
upper=np.where(df['sepal.width']>=(Q3+1.5*IQR))
```

```python
lower=np.where(df['sepal.width']<=(Q1-1.5*IQR))
```

```python
df.drop(upper[0],inplace=True)
```

```python
df.drop(lower[0],inplace=True)
```

```python
df.shape
```
