```python
import pandas as pd
```

```python
import numpy as nm
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['age']=df['age'].replace(nm.nan,df['age'].mean())
```

```python
sns.boxplot(x='sex',y='age',data=df)
```

```python
sns.boxplot(x='age',data=df)
```

```python
Q1=nm.percentile(df['age'],25,interpolation='midpoint')
```

```python
print(Q1)
```

```python
Q3=nm.percentile(df['age'],75,interpolation='midpoint')
```

```python
print(Q3)
```

```python
IQR=Q3-Q1
```

```python
df.shape
```

```python
upper=nm.where(df['age']>=(Q3+0.1*IQR))
```

```python
lower=nm.where(df['age']<=(Q1-0.1*IQR))
```

```python
df.drop(upper[0],inplace=True)
```

```python
df.drop(lower[0],inplace=True)
```

```python
df.shape
```

```python
sns.boxplot(x='age', data=df)
```
